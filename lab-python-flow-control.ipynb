3. Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for  the products that were ordered (those in "customer_orders").

```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
# print ("Type 'Exit' at any time to stop purchasing")
print ("Available products", products)

inventory = {}
for product in products:
    quantity = int(input(f"Enter quantity of {product}: ").strip())
    inventory[product] = quantity

customer_orders = set()

while True:
    order = input('Enter the name of the product to purchase: ').strip().lower()
    
    if order in products:
        customer_orders.add(order)
        
    else:
        while order not in products:
            print (f"'{order}' is not in our inventory. Please select another item.")
            order = input('Enter the name of the product to purchase: ').strip().lower()
            if order in products:
                customer_orders.add(order)

    validation = input("Do you want to buy more items? (yes/no): ").strip().lower()

    while validation not in ("no", "yes"):
        print(f"'{validation}' is not a correct option. Please select 'yes' or 'no'.")
        validation = input("Do you want to buy more items? (yes/no): ").strip().lower()

    if validation == "no":
        break
    

print (f"The customer has ordered {customer_orders}" "\n")

total_products_ordered = len(customer_orders)
percentage_ordered = (total_products_ordered/len(products))*100

order_status = (total_products_ordered, percentage_ordered)

print ("Order Statistics:" "\n", 
f"Total Products Ordered: {order_status[0]}" "\n",
f"Percentage of Products Ordered : {order_status[1]}%" "\n")

for item in customer_orders:
        inventory[item] -= 1

for key,value in inventory.items():
    print (f"The actual quantity of item {key} is: {value}")
```
